```python
class Solution(object):
    def maxArea(self, height):
        """
        :type height: List[int]
        :rtype: int
        :pos1, 一条边的index
        :pos2, 一条边的index
        :height[pos1], 边高
        :height[pos2], 边高
        :eq = min(height[pos1],height[pos2])*(pos2-pos1), 该如何使得eq最大化
        for 循环双重遍历会超出时间范围
        假设pos1=0，pos2会遍历，那么永远都会被限制在height[pos1]的位置上

        应该从中间开始找：
            pos1 = 0
            pos2 = n
            eq = min(height[pos1],height[pos2])*(pos2-pos1)
            设置cur_pos1, cur_pos2
            cur_pos1 = pos1
            cur_pos2 = pos2
            那么移动指针就有3种情况
            1. cur_pos1+1, cur_pos2,    eq1
            2. cur_pos1, cur_pos2-1,    eq2
            3. cur_pos1+1, cur_pos2-1,  eq3
            连续比较eq与eq1，eq2，eq3的大小，如果有的话，则改进
            循环结束条件：pos1 == pos2, end cycle
        """
        left_p = len(height)//2
        right_p = len(height)//2

        cur_p1 = left_p
        cur_p2 = right_p

        max_eq = min(height[left_p],height[right_p])*(right_p - left_p)
        cur_eq = max_eq
        # pos1 会移动， pos2也会移动
        while left_p != right_p:
            # 第一种情况，left 指针
            eq1 = min(height[left_p+1],height[right_p])*(right_p - left_p - 1)
            if cur_eq <= eq1:
                cur_eq = eq1
                cur_p1 = left_p + 1
                cur_p2 = right_p

            eq2 = min(height[left_p],height[right_p-1])*(right_p - left_p - 1)
            if cur_eq <= eq2:
                cur_eq = eq2
                cur_p1 = left_p
                cur_p2 = right_p - 1
            
            eq3 = min(height[left_p+1],height[right_p-1])*(right_p - left_p - 2)
            if cur_eq <= eq3:
                cur_eq = eq3
                cur_p1 = left_p + 1
                cur_p2 = right_p - 1      

            

        
```

## 算法核心思想：如何更新两个指针

是否更新指针，以及如何更新指针，需要考虑公式 

$$
Area = min(height[pos_{left}],height[pos_{right}])*(pos_{right} - pos_{left})
$$

因为 pos_left 和 pos_right 是双向移动的，两者之间的距离是在逐渐减小的，这就给选择往哪边移动提供了判断标准：选择把两者之间小的那个方向移动


```python
height = [1,8,6,2,5,4,8,3,7]

N = len(height)
left = 0
right = N-1
max_area = 0

while left != right:
    area = min(height[left], height[right])*(right-left)
    max_area = max(max_area,area)

    # 判断指针固定移动方向
    if height[left] < height[right]:
        left+=1
    else:
        right-=1

max_area
```
