# 两数之和
给定一个整数数组 ·`nums` 和一个整数目标值 `target`，请你在该数组中找出和为目标值 `target` 的那两个整数，并返回它们的数组下标。

你可以假设每种输入只会对应一个答案，并且你不能使用两次相同的元素。

你可以按任意顺序返回答案。

示例1：
```
输入：nums = [2,7,11,15], target = 9
输出：[0,1]
解释：因为 nums[0] + nums[1] == 9 ，返回 [0, 1] 。
```

示例2:
```
输入：nums = [3,2,4], target = 6
输出：[1,2]
```

## 方法1：暴力求解，使用双层for循环
时间复杂度：$O(n^2)$

````python
import time
class Solution(object):
    def twoSum(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: List[int]
        """
        N = target
        a = 0
        b = 0
        for i in range(len(nums)):
            a = nums[i]
            for j in range(len(nums)):
                if i == j:
                    continue
                else:
                    b = nums[j]
                    if a+b == N:
                        return [i,j]

sample = Solution()
nums = [2,7,11,15]
target = 9

start = time.time()
x = sample.twoSum(nums, target)
end = time.time()

print(x)
print(end-start)
````

## 方法二：使用Hash函数
使用一个dict存储为nums中的数据为{num:id}，时间复杂度$O(n)$，然后使用for循环遍历dict的key，查找key的补值是否存在，若存在就返回

````python
class Solution(object):
    def twoSum(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: List[int]
        """
        a = 0
        b = 0
        num_to_index = dict()
        for i in range(len(nums)):
            num_to_index[nums[i]] = i

        for i in range(len(nums)):
            a = nums[i]
            b = target - a
            if num_to_index.get(b) is not None:
                if num_to_index[b] != i:
                    return [i, num_to_index[b]]

sample = Solution()
nums = [2,7,11,15]
target = 9

start = time.time()
x = sample.twoSum(nums, target)
end = time.time()

print(x)
print(end-start)
````

的确，时间复杂度上方案2比方案1要少接近4倍
